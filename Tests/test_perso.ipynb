```python
import sys
import os

chemin_actuel = os.getcwd()
chemin_parent = os.path.dirname(chemin_actuel)
sys.path.append(chemin_parent)

from infos_perso import *
from class_sportif import *
```

```python
def test_func_infos_personelles():
    sportif_test = func_infos_personelles(
        nom="Dupont",
        prenom="Marie",
        sexe="Femme",
        age=26,
        nationalite="Française",
        localisalisation=[48.8566, 2.3522],
        distance_rencontre=10,
        niveau_sports={"yoga": "avancé", "course": "intermédiaire"},
        attentes=["bien-être", "motivation"],
        genre_recherche="Homme",
        min_age_recherchee=24,
        max_age_recherchee=32
    )

    assert sportif_test.nom == "Dupont", "Le nom est incorrect"
    assert sportif_test.age == 26, "L'âge n'est pas correctement enregistré"
    assert isinstance(sportif_test.niveau_sports, dict), "Le champ 'niveau_sports' doit être un dictionnaire"
    assert "course" in sportif_test.niveau_sports, "La clé 'course' devrait être dans niveau_sports"
    assert 24 <= sportif_test.min_age_recherchee <= sportif_test.max_age_recherchee, "Les bornes d'âge sont incohérentes"

    print("Tous les tests sont passés avec succès !")
    print(sportif_test)


# Exécution du test
test_func_infos_personelles()
```
